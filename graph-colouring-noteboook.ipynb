Introduction to Vertex Colouring
===============================

This notebook is about vertex colouring of graphs.

Sequential Vertex Colouring
---------------------------

The simplest way to colour vertices is ... $x^{2}$.

```python
import networkx as nx
def setfigsize(n, m):
  pylab.rcParams['figure.figsize'] = n, m
setfigsize(5, 5)
```

```python
def firstavailable(G, node):
    """Returns the first colour in [1,,D+1] which is not used in G on any
    neighbours of node, where D is the maximum degree."""
    palette = range(1, G.number_of_nodes() + 1)
    used_on_neighbours = []
    for v in G[node]:
        used_on_neighbours.append(G.node[v].get('colour'))
    available_colour_set = set(palette) - set(used_on_neighbours)
    return sorted(available_colour_set)[0]

def vcolour(G, strategy=firstavailable):
    """Visits every vertex in G and assigns a colour given by the strategy
    object."""
    for node in G.node:
        G.node[node]['colour'] = strategy(G, node)
```

```python
def colours(G):
    """Returns a list of colours used on vertices in G."""
    return [x.get('colour') for x in G.node.values()]

def ncolours(G):
    """Counts the number of different colours used on vertices in G."""
    return len(set(colours(G)))   
```

```python
def standard_colours(G):
    """Appears to do the same as a later function called colours."""
    return [G.node[node]['colour'] for node in G.node.keys()]
```

```python
options = {
 'with_labels': False,
 'node_size': 300,
 'width': 1,
}
```

```python
K9 = nx.complete_graph(9)
vcolour(K9)
nx.draw_circular(K9, node_color = standard_colours(K9), **options)
ncolours(K9)
```

```python
K56 = nx.complete_bipartite_graph(5,6)
vcolour(K56)
nx.draw_circular(K56, node_color = standard_colours(K56), **options)
ncolours(K56)
```

## Colouring some classic graphs

```python
P = nx.petersen_graph()
vcolour(P)
nx.draw_shell(P, nlist=[range(5,10), range(5)], node_color = standard_colours(P), **options)
ncolours(P)
```

```python
setfigsize(6, 6)

G = nx.dodecahedral_graph()
vcolour(G)
nx.draw_shell(G, nlist = [[2,3,4,5,6],[8,1,0,19,18,17,16,15,14,7],[9,10,11,12,13]], node_color = standard_colours(G), **options)
```
