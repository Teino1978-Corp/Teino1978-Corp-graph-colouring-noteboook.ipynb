Introduction to Vertex Colouring
===============================

This notebook is about vertex colouring of graphs.

1. Greedy Colouring
-------------------

In this section we introduction greedy colouring.

```python
import networkx as nx
```

```python
def firstavailable(G, node):
    palette = range(1, G.number_of_nodes() + 1)
    used_on_neighbours = []
    for v in G[node]:
        used_on_neighbours.append(G.node[v].get('colour'))
    available_colour_set = set(palette) - set(used_on_neighbours)
    return sorted(available_colour_set)[0]

def vcolour(G, strategy=firstavailable):
    for node in G.node:
        G.node[node]['colour'] = strategy(G, node)
```

```python
def standard_colours(G):
    return [G.node[node]['colour'] for node in G.node.keys()]
```

```python
G = nx.complete_graph(9)
vcolour(G)
nx.draw_circular(G, node_color = standard_colours(G))
```

```python
G = nx.complete_bipartite_graph(5,6)
vcolour(G)
nx.draw_circular(G, node_color = standard_colours(G))
```

```python
G = nx.petersen_graph()
vcolour(G)
nx.draw_circular(G, node_color = standard_colours(G))
```

```python
G = nx.complete_graph(10)
vcolour(G)
P = nx.drawing.circular_layout(G, scale = 2.5)
D = nx.draw(G, pos = P, node_color = standard_colours(G))
import matplotlib.pyplot as plt
plt.savefig("graph.png")
```
